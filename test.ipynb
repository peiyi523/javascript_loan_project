```python
period = years * 12
month_pay = int(amount / period)
month_pay
```

```python
# 第一個月利息
month_rate = rate / 100 / 12
```

```python
month_rate
```

每月還款金額及利息


```python
amount = 100000
years = 5
rate = 2
period = years * 12
month_rate = rate / 100 / 12
datas = []
for i in range(period):
    interest = round(amount * month_rate)
    amount -= month_pay
    if i == period - 1:
        datas.append(
            [i + 1, month_pay + amount, interest, month_pay + amount + interest, 0]
        )
        print(
            f"期別:{i+1} 本金:{month_pay+amount} 利息:{interest} 本利和:{month_pay+interest+amount} 餘額:{0}"
        )
    else:
        datas.append([i + 1, month_pay, interest, month_pay + interest, amount])
        print(
            f"期別:{i+1} 本金:{month_pay} 利息:{interest} 本利和:{month_pay+interest} 餘額:{amount}"
        )
```
