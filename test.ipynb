本息攤還

```python
def rule2(total_amount, years, rate):
    amount = total_amount
    months = years * 12
    month_rate = rate / 100 / 12

    # 本息平均攤還每月還款金額
    month_pay = amount * (month_rate * pow(1 + month_rate, months)) / (
                pow(1 + month_rate, months) - 1)

    totalInterest = 0
    datas = []

    for i in range(period):
        interest = round(amount * month_rate)
        principal_payment = month_pay - interest
        amount -= principal_payment

        if i == period - 1:
            datas.append([i + 1, month_pay + amount, interest, month_pay + amount + interest, 0])
        else:
            datas.append([i + 1, month_pay, interest, month_pay + interest, amount])

        totalInterest += interest

    return [datas, totalInterest]

# 示例用法
total_amount = 100000  # 贷款总额
loan_years = 5  # 贷款年限
interest_rate = 2  # 年利率（百分比）

schedule_result, total_interest = rule2(total_amount, loan_years, interest_rate)


```
