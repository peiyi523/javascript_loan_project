本息攤還

```python
def rule2(total_amount, years, rate):
    amount = total_amount
    period = years * 12
    month_rate = rate / 100 / 12
    # 本息平均攤還每月還款金額
    month_pay = round(
        (
            amount
            * (month_rate * pow(1 + month_rate, period))
            / (pow(1 + month_rate, period) - 1)
        ),
        0,
    )

    totalInterest = 0
    datas = []

    for i in range(period):
        interest = round(amount * month_rate)
        principal_payment = month_pay - interest
        amount -= principal_payment

        if i == period - 1:
            datas.append(
                [i + 1, month_pay + amount, interest, month_pay + amount + interest, 0]
            )
        else:
            datas.append([i + 1, month_pay, interest, month_pay + interest, amount])

        totalInterest += interest

    return [datas, totalInterest]
```

```python
total_amount = 100000
loan_years = 5
interest_rate = 2

schedule_result, totalInterest = rule2(total_amount, loan_years, interest_rate)
schedule_result, totalInterest
```
